```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
data=pd.read_csv("advertising.csv")
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
d=data.drop(['Unnamed: 0'],axis=1,inplace=True)
data
```

```python
x=np.array(data.drop(["Sales"], 1))
y=np.array(data["Sales"])
```

```python
xtrain, xtest, ytrain, ytest= train_test_split(x, y,train_size=0.8,test_size=0.2,random_state=0)
```

```python
model= LinearRegression()
model.fit(xtrain, ytrain)
```

```python
yprediction = model.predict(xtest)
yprediction
```

```python
r2_score(ytest,yprediction)
```

```python
pred_sales = pd.DataFrame(data={"Predicted Sales": yprediction.flatten()})
```
